```python
import tensorflow as tf
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mimg
from tensorflow.keras.layers import Dense
import zipfile
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
```

```python
files = zipfile.ZipFile('/content/Diabetes Datadet.zip')
files.extractall()
```

```python
dataset = pd.read_csv('/content/diabetes_prediction_dataset.csv')
```

```python
df = pd.DataFrame(dataset)
df.head()
```

```python
df.tail()
```

```python
## There are some values 'NO INFO' in smoking histoty we actually Dont't need those data. So we are just getting the data that are not have 'No Info' in smoking history

df = df[df['smoking_history']!='No Info']
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
## Converting the Gender column into 0's and 1's
## Male -- 1 and Female --0

le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
df.head()
```

```python
df.tail()
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values

x = np.array(x)
```

```python
df['smoking_history'].unique()
```

```python
## Applying OneHotEncoding for the Smoking History column. Because it having more than 4 categories

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

cl = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[4])],remainder='passthrough')
x = np.array(cl.fit_transform(x))
```

```python
x = x.astype(np.float32)
y = y.astype(np.float32)
```

```python
x[0]
```

```python
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
type(x_train)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model_1 = Sequential([
    Dense(10,activation='relu'),
    Dense(10,activation='relu'),

    Dense(1,activation='sigmoid')
])

model_1.compile(loss='binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])

history_1 = model_1.fit(
    x_train,y_train,
    validation_data = (x_test,y_test),
    epochs = 10,
    batch_size = 32
)
```

```python
model_1.evaluate(x_test,y_test)
```

```python
lis2 = ['Female',72.0,0,1,'former',27.94,6.5,130]

lis2 = np.array(lis2)

lis2[0] = le.transform([lis2[0]])[0]

lis2 = lis2.reshape(1,-1)
lis2 = cl.transform(lis2)
lis2 = lis2.astype('float32')
lis2

pred = model_1.predict(lis2)
print(pred)

if pred[0][0] > 0.5:
  print("Diabetes")
else:
  print("Not Diabetes")
```
