```python
import tensorflow as tf
import os
import matplotlib.pyplot as plt
import matplotlib.image as mimg
import zipfile
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
files = zipfile.ZipFile('Cancer dataset.zip')
files.extractall()
```

```python
dataset = pd.read_csv('/content/Cancer_Data.csv')
```

```python
df = pd.DataFrame(dataset)
df.head()
```

```python
df.tail()
```

```python
df.drop('id',inplace=True,axis='columns')
```

```python
df.head()
```

```python
df.drop('Unnamed: 32',inplace=True,axis='columns')
```

```python
df.head()
```

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head()
```

```python
x = df.drop('diagnosis',axis='columns').values
y = df.iloc[:,0].values
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
sc = StandardScaler()
x_train_standardized = sc.fit_transform(x_train)
x_test_standardized = sc.transform(x_test)
```

```python
model_1 = Sequential([
    Dense(10,activation='relu'),
    Dense(10,activation='relu'),

    Dense(1,activation='sigmoid')
])

model_1.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

history_1 = model_1.fit(
    x_train_standardized,y_train,
    validation_data = (x_test_standardized,y_test),
    epochs = 10,
    batch_size = 32
)
```

```python
model_1.evaluate(x_test_standardized,y_test)
```

```python
## Model Prediction

lis = [12.05,14.63,78.04,449.3,0.1031,0.09092,0.06592,0.02749,0.1675,0.06043,0.2636,0.7294,1.848,19.87,0.005488,0.01427,0.02322,0.00566,0.01428,0.002422,13.76,20.7,89.88,582.6,0.1494,0.2156,0.305,0.06548,0.2747,0.08301]
lis = np.array(lis)
lis = lis.reshape(1,-1)
lis = sc.transform(lis)
pred = model_1.predict(lis)
print(pred)

if pred[0][0] >= 0.5:
  print("Cancer")
else:
  print("No Cancer")
```
